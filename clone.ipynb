WQI generation

```python
df['Qc'] = df['Conductivity']/5
df['Qn'] = df['NO3']/0.5
df['Qtemp'] = df['Temp']/0.15
df['Qtu'] = df['Turbidity']/0.05
```

```python
df['Wc'] = df['Qc']*0.006928422
df['Wn'] = df['Qn']*0.06928422
df['WTemp'] = df['Qtemp']*0.2309474
df['Wtu'] = df['Qtu']*0.6928422
df.head()
```

```python
df['Index']= df['Wc']+df['Wn']+df['WTemp']+df['Wtu']
print(df.head())
```

```python
#x=dff.drop("Index",axis=1) #features
col_to_drop=["Timestamp","Q","Level", "Dayofweek", "Month", "Qc","Qn","Qtemp","Qtu","Wc","Wn","WTemp","Wtu","Index"]
x=df.drop(columns=col_to_drop,axis=1)
y=df["Index"] #index
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix #for classification model
from sklearn.metrics import mean_absolute_error,mean_squared_error #for regression model
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3) #splitting data for 70% training and 30% testing
x_train.shape,x_test.shape
```

K-fold cross validation for decision tree

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor


model_dt = DecisionTreeRegressor()

# Define the range of k values
k_values = range(2, 20)

# Perform k-fold cross-validation for each k value
for k in k_values:
    # Perform k-fold cross-validation
    scores = cross_val_score(model_dt, x_train, y_train, cv=k, scoring='neg_mean_absolute_error')

    # Calculate mean absolute error across all folds
    mae_dt_cross_val = -scores.mean()

    print("Mean Absolute Error (Cross-Validated) for k =", k, ":", mae_dt_cross_val)
```

Cross validation on Random Forest

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Define the range of values for the number of folds (k)
k_range = range(2, 6)

# Initialize an empty list to store the mean absolute errors
mae_scores = []

# Initialize Random Forest Regressor
model_rf = RandomForestRegressor()

# Iterate over the range of k values
for k in k_range:
    # Perform k-fold cross-validation
    scores = cross_val_score(model_rf, x_train, y_train, cv=k, scoring='neg_mean_absolute_error')

    # Convert scores to positive values
    mae = -scores.mean()

    # Append the mean absolute error to the list
    mae_scores.append(mae)

# Print the mean absolute errors for each value of k
for k, mae in zip(k_range, mae_scores):
    print(f"Mean Absolute Error (k={k}): {mae}")
```

K fold cross validation on knn

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error

# Define a range of  values
k_range = range(2, 21)

# Initialize an empty list to store mean absolute errors for each neighbor value
mae_scores = []

# Iterate over the range
for k in k_range:
    # Initialize K-Nearest Neighbors Regressor
    model_knn = KNeighborsRegressor(n_neighbors=5)

    # Perform k-fold cross-validation
    scores = cross_val_score(model_knn, x_train, y_train, cv=k, scoring='neg_mean_absolute_error')

    # Convert scores to positive values and calculate mean absolute error
    mae_knn_cv = -scores.mean()

    # Append mean absolute error to the list
    mae_scores.append(mae_knn_cv)

# Print mean absolute errors for each neighbor value
for k, mae in zip(k_range, mae_scores):
    print(f"Mean Absolute Error (Cross-Validated) for {k} value:", mae)
```

k fold cross validation for Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error

# Define a range of folds
fold_range = range(2, 21)

# Initialize an empty list to store mean absolute errors for each fold value
mae_scores = []

# Initialize Linear Regression model
model_lr = LinearRegression()

# Iterate over the range of folds
for fold in fold_range:
    # Perform k-fold cross-validation
    mae_lr_cv = -cross_val_score(model_lr, x_train, y_train, cv=fold, scoring='neg_mean_absolute_error').mean()

    # Append mean absolute error to the list
    mae_scores.append(mae_lr_cv)

# Print mean absolute errors for each fold value
for fold, mae in zip(fold_range, mae_scores):
    print(f"Mean Absolute Error (Cross-Validated) for {fold} folds:", mae)
```
